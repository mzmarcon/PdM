```python
import scipy.io
import pandas as pd
import numpy as np

mat = scipy.io.loadmat('mill.mat')
df = pd.DataFrame(mat['mill'][0])
df.drop(index=94)
df
```

```python
sequence_df = []
for column in df.columns[:7]:
    sequence_list = []
    for item in df[column]:
        sequence_list.append(item.item())
    sequence_df.append(np.array(sequence_list))
```

```python
dataset = pd.DataFrame(np.array(sequence_df).T)
dataset.columns = df.columns[:7]
dataset = dataset.join(df.iloc[:,7:])
```

```python
dataset
```

```python
df_explode = dataset.apply(lambda x: x.explode() if x.name in dataset.columns[7:] else x)
```

```python
df_explode.smcAC.map(lambda x: x.item()).plot()
```
